```python
#loadin the necessary libraries/imports
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from os import listdir
from keras.preprocessing import sequence
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.optimizers import Adam
from keras.models import load_model
from keras.callbacks import ModelCheckpoint
```

```python
#reading all the files from the folder and putting it in a sequence list.(all 216 files)
path = 'DMV_CS551G_Assessment1_Task2_dataset/Ambient_Living_'
sequences = list()
for i in range(1,217):
    file_path = path + str(i) + '.csv'
    print(file_path)
    df = pd.read_csv(file_path, header=0)
    values = df.values
    sequences.append(values)
```

```python
#reading the target file
targets = pd.read_csv('DMV_CS551G_Assessment1_Task2_dataset/Ambient_Living_target.csv')
targets = targets.values[:,1] #taking the target labels
```

```python
#cheking the key information about the sequence list
#diffrent data file have diffrent lengh.this will give us a overall view of the dataets
len_sequences = []
for one_seq in sequences:
    len_sequences.append(len(one_seq))
pd.Series(len_sequences).describe()
```

```python
#now we can chck the 80th quartile to decide the lenght of the files we want for this report
q3=pd.Series(len_sequences).quantile(.80)
print("The length of the files should be :",q3 ,"to comply with 80th Quartile")
```

```python
#returning the length of each sequence in the list
#len_sequences
```

```python
#Padding the sequence with 0 to max length
to_pad = 103
new_seq = []
for one_seq in sequences:
    len_one_seq = len(one_seq)
    last_val = one_seq[-1]
    n = to_pad - len_one_seq
    #to_concat = np.repeat(one_seq[-1], n).reshape(4, n).transpose()
    to_concat = np.repeat([0,0,0,0], n).reshape(4, n).transpose()
    new_one_seq = np.concatenate([one_seq, to_concat])
    new_seq.append(new_one_seq)
final_seq = np.stack(new_seq)
```

```python
#final_seq[1]
```

```python
#truncate the sequence to length 54
from keras.preprocessing import sequence
seq_len = 54
final_seq=sequence.pad_sequences(final_seq, maxlen=seq_len, padding='post', dtype='float', truncating='post')
```

```python
len(final_seq)
```

```python
#final_seq
```

#### Test and train split

```python
np.random.seed(31416) 
#creating test and train sets
train = [final_seq[i] for i in range(0,180)]
test = [final_seq[i] for i in range(180,216)]

train_target = [targets[i] for i in range(0,180)]
test_target = [targets[i] for i in range(180,216)]
len(train)
```

```python
len(test)
```

```python
#train_target
```

```python
#test_target
```

#### Test and train target data transformation

```python
#changeing the label values to 0 and 1
np.random.seed(31416)
train = np.array(train)
test = np.array(test)
train_target = np.array(train_target)
train_target = (train_target+1)/2

test_target = np.array(test_target)
test_target = (test_target+1)/2
```

```python
test_target
```

### LSTM base_model

```python
np.random.seed(31416)
#creating LSTM model
model = Sequential()
model.add(LSTM(256, input_shape=(seq_len, 4)))
model.add(Dense(1, activation='sigmoid'))
#define and optimise LSTM trainning process
#creating checkpoit to get outcome of each epoches
adam = Adam(lr=0.001)
chk = ModelCheckpoint('best_model.pkl', monitor='val_acc', save_best_only=True, mode='max', verbose=1)
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
model.fit(train, train_target, epochs=200, batch_size=128, callbacks=[chk], validation_data=(test,test_target))
```

```python
np.random.seed(31416)
#loading the model and checking accuracy on the test data
model = load_model('best_model.pkl')
from sklearn.metrics import accuracy_score
test_preds = model.predict_classes(test)
print("Accuracy of the model is :",accuracy_score(test_target, test_preds))
```

```python
score,acc = model.evaluate(test, test_target)
print('Test score:', score)
print('Test accuracy:', acc)
```

```python
model.summary()
```

#### 10 fold cross validation for base model

```python
from sklearn.model_selection import KFold
from pylab import *
from sklearn.metrics import accuracy_score
np.random.seed(31416)
scores = [] 
cv = KFold(n_splits=10)
X=np.array(final_seq)
Y=np.array(targets)
for train_index, test_index in cv.split(X,Y):
    X_train=X[train_index]
    X_test=X[test_index]
    y_train=(Y[train_index]+1)/2
    y_test=(Y[test_index]+1)/2
    model = Sequential()
    model.add(LSTM(256, input_shape=(seq_len, 4)))
    model.add(Dense(1,activation='sigmoid'))
    adam = Adam(lr=0.001)
    chk = ModelCheckpoint('best_model.pkl', monitor='val_acc', save_best_only=True, mode='max', verbose=1)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    model.fit(X_train, y_train, epochs=200, batch_size=128, callbacks=[chk], validation_data=(X_test,y_test))
    model = load_model('best_model.pkl')
    test_preds = model.predict_classes(X_test)
    scores.append(accuracy_score(y_test, test_preds))
    print("------------------------------------------------------")
```

```python
print(scores)
np.mean(scores)
```
